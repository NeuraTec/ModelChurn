```python
#================== LIBRER√çAS ==================
import warnings
warnings.filterwarnings("ignore")

import logging
from logging.handlers import RotatingFileHandler

from sklearn.metrics import roc_auc_score
import shap
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_predict
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score, fbeta_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.feature_selection import SelectFromModel, VarianceThreshold
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
import matplotlib.pyplot as plt
import joblib
import os
from datetime import datetime

# ================== CONFIGURACI√ìN GLOBAL ==================
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# ================== CONFIGURACI√ìN DE LOGGING ==================
logger = logging.getLogger()
logger.setLevel(logging.INFO)

# Evitar duplicados eliminando handlers previos
if logger.hasHandlers():
    logger.handlers.clear()

# Directorio logs (crearlo si no existe)
os.makedirs("logs", exist_ok=True)

# --- Handler para archivo ---
file_handler = RotatingFileHandler(
    "logs/entrenamiento.log",
    maxBytes=5_000_000,
    backupCount=3,
    encoding="utf-8"
)
file_format = logging.Formatter(
    "%(asctime)s - %(levelname)s - %(message)s"
)
file_handler.setFormatter(file_format)
logger.addHandler(file_handler)

# --- Handler para consola ---
console_handler = logging.StreamHandler()
console_format = logging.Formatter("%(levelname)s - %(message)s")  # Consola m√°s limpia
console_handler.setFormatter(console_format)
logger.addHandler(console_handler)

# ================== FUNCIONES ==================
def cargar_y_preprocesar_datos(ruta_csv):
    """Carga el dataset, limpia columnas innecesarias y prepara los datos."""
    logger.info("Cargando dataset desde: %s", ruta_csv)
    data = pd.read_csv(ruta_csv)
    data
    data.drop('customerID', axis=1, inplace=True)
    data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
    data.dropna(subset=['TotalCharges'], inplace=True)
    X = data.drop('Churn', axis=1)
    y = data['Churn']
    logger.info("Dataset cargado correctamente. Filas: %d, Columnas: %d", data.shape[0], data.shape[1])
    return X, y


def dividir_datos(X, y):
    """Divide los datos en entrenamiento y prueba y codifica las etiquetas."""
    logger.info("Dividiendo datos en train y test...")
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=RANDOM_STATE)
    label_enc = LabelEncoder()
    y_train_enc = label_enc.fit_transform(y_train)
    y_test_enc = label_enc.transform(y_test)
    logger.info("Datos divididos: Train=%d, Test=%d", len(y_train_enc), len(y_test_enc))
    return X_train, X_test, y_train_enc, y_test_enc, label_enc


def construir_preprocesador(X):
    """Crea un transformador que escala variables num√©ricas y codifica variables categ√≥ricas."""
    logger.info("Construyendo preprocesador (StandardScaler + OneHotEncoder)...")
    num_cols = X.select_dtypes(include=['float64', 'int', 'int64']).columns.tolist()
    cat_cols = X.select_dtypes(include='object').columns.tolist()
    logger.info("Columnas num√©ricas: %d | Columnas categ√≥ricas: %d", len(num_cols), len(cat_cols))
    preprocessor = ColumnTransformer([
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ])
    return preprocessor
```

```python
def make_pipeline_with_fs(model, preprocessor, use_fs=False, fs_threshold='median'):
    """Construye un pipeline con preprocesamiento, selecci√≥n de caracter√≠sticas opcional, balanceo y modelo final."""
    steps = [
        ('preprocessor', preprocessor),
        ('low_var_filter', VarianceThreshold(threshold=0.01))
    ]

    if use_fs:
        sel = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE), threshold=fs_threshold)
        steps.append(('feature_selection', sel))

    steps.append(('smote', SMOTE(random_state=RANDOM_STATE)))
    steps.append(('model', model))
    return ImbPipeline(steps)


def mostrar_features_seleccionados(pipeline, X, modelo_nombre):
    """Extrae correctamente las features seleccionadas siguiendo las transformaciones reales."""
    if 'feature_selection' not in pipeline.named_steps:
        logger.warning("‚ö†Ô∏è No existe etapa de selecci√≥n de caracter√≠sticas en este pipeline.")
        return

    steps = pipeline.named_steps

    # === 1) Preprocesamiento ===
    X_pre = steps['preprocessor'].transform(X)

    # === 2) Variance threshold ===
    X_low = steps['low_var_filter'].transform(X_pre)

    # === Obtener nombres correctos ===
    try:
        feature_names = steps['preprocessor'].get_feature_names_out()
        feature_names = feature_names[steps['low_var_filter'].get_support()]
    except Exception:
        feature_names = np.arange(X_low.shape[1]).astype(str)

    # === 3) Feature selection ===
    selector = steps['feature_selection']
    mask = selector.get_support()

    if len(mask) != len(feature_names):
        logger.error("‚ùå ERROR: Tama√±o del mask no coincide con el de las features.")
        logger.error("Features: %d, Mask: %d", len(feature_names), len(mask))
        return

    selected_features = np.array(feature_names)[mask]

    # === Guardar ===
    df = pd.DataFrame(selected_features, columns=['feature'])
    archivo = f"features_seleccionadas_{modelo_nombre}.csv"
    df.to_csv(archivo, index=False)

    logger.info("‚úÖ Features seleccionadas: %d", len(selected_features))
    logger.info("üìÅ Guardado en %s", archivo)


def entrenar_modelos(X_train, y_train, preprocessor, cv, fs_threshold='median', force_fs=True, scoring='recall'):
    """Entrena y eval√∫a varios modelos con validaci√≥n cruzada y b√∫squeda de hiperpar√°metros."""
    models = {
        'logistic': {
            'model': LogisticRegression(solver='liblinear', random_state=RANDOM_STATE),
            'param_grid': {'model__C': [0.001, 0.01, 0.1, 1, 10], 'model__penalty': ['l2']}
        },
        'logistic_lasso': {
            'model': LogisticRegression(solver='liblinear', penalty='l1', random_state=RANDOM_STATE),
            'param_grid': {'model__C': [0.01, 0.1, 1, 10]}
        },
        'lgbm': {
            'model': LGBMClassifier(random_state=RANDOM_STATE, verbosity=-1),
            'param_grid': {
                'model__n_estimators': [100],
                'model__max_depth': [3, 5],
                'model__learning_rate': [0.01, 0.1],
                'model__num_leaves': [15, 31]
            }
        },
        'rf': {
            'model': RandomForestClassifier(random_state=RANDOM_STATE),
            'param_grid': {
                'model__n_estimators': [100, 200],
                'model__max_depth': [None, 10],
                'model__min_samples_split': [2, 5]
            }
        }
    }

    resultados = {}
    logger.info("üîÑ Ajustando modelos y buscando mejores hiperpar√°metros...")

    for name, info in models.items():
        model = info['model']
        param_grid = info['param_grid']
        fs_options = [False] if (not force_fs or name == 'logistic_lasso') else [False, True]

        for use_fs in fs_options:
            tag = f"{name}{'_fs' if use_fs else ''}"
            logger.info("\n‚û°Ô∏è Pipeline: %s", tag)

            pipeline = make_pipeline_with_fs(model, preprocessor, use_fs=use_fs, fs_threshold=fs_threshold)
            grid = GridSearchCV(pipeline, param_grid, scoring=scoring, cv=cv, n_jobs=-1)
            grid.fit(X_train, y_train)

            logger.info(" Mejores par√°metros: %s", grid.best_params_)
            logger.info(" Mejor recall CV: %.4f", round(grid.best_score_, 4))

            resultados[tag] = {
                'grid': grid,
                'score': grid.best_score_,
                'model': grid.best_estimator_
            }
    return resultados
```

```python
def explicar_modelo_final_shap(pipeline, X, modelo_nombre="modelo"):
    """Genera SHAP usando exactamente las mismas transformaciones que vio el modelo."""
    logger.info("\nüîç Generando explicaciones SHAP para el modelo final...")

    steps = pipeline.named_steps

    # === 1) Preprocesador ===
    X_t = steps['preprocessor'].transform(X)

    # === 2) Variance Threshold ===
    X_t = steps['low_var_filter'].transform(X_t)

    # === 3) Feature Selection (opcional) ===
    if 'feature_selection' in steps:
        X_t = steps['feature_selection'].transform(X_t)

    # === Obtener nombres reales ===
    try:
        feature_names = steps['preprocessor'].get_feature_names_out()
        feature_names = feature_names[steps['low_var_filter'].get_support()]

        if 'feature_selection' in steps:
            feature_names = feature_names[steps['feature_selection'].get_support()]

    except Exception:
        feature_names = np.arange(X_t.shape[1]).astype(str)

    # === 4) Modelo final ===
    model_final = steps['model']

    # === 5) SHAP ===
    explainer = shap.Explainer(model_final, X_t)
    shap_values = explainer(X_t)

    # === 6) Gr√°fico ===
    plt.figure()
    shap.summary_plot(
        shap_values,
        features=X_t,
        feature_names=feature_names,
        show=False
    )
    plt.tight_layout()
    plt.savefig(f"shap_summary_{modelo_nombre}_final.png", bbox_inches='tight', dpi=300)
    plt.show()
    plt.close()

    logger.info("‚úÖ SHAP guardado: shap_summary_%s_final.png", modelo_nombre)


def ajustar_threshold(best_model, X_train, y_train, beta=2.0):
    """Ajusta el umbral de decisi√≥n del modelo y eval√∫a m√©tricas."""
    logger.info("\nüìä Ajustando threshold de decisi√≥n...")
    y_probs = cross_val_predict(best_model, X_train, y_train, cv=5, method='predict_proba')[:, 1]

    thresholds = np.arange(0.2, 0.61, 0.05)
    resultados = []

    for t in thresholds:
        y_pred = (y_probs >= t).astype(int)
        resultados.append({
            'Threshold': round(t, 2),
            'Precision': round(precision_score(y_train, y_pred), 3),
            'Recall': round(recall_score(y_train, y_pred), 3),
            'F1-score': round(f1_score(y_train, y_pred), 3),
            f'F{int(beta)}-score': round(fbeta_score(y_train, y_pred, beta=beta), 3)
        })

    df = pd.DataFrame(resultados)
    df.to_csv("tabla_thresholds.csv", index=False)
    logger.info("‚úÖ Tabla de thresholds guardada como 'tabla_thresholds.csv'")

    plt.figure(figsize=(10, 6))
    for col in ['Precision', 'Recall', 'F1-score', f'F{int(beta)}-score']:
        plt.plot(df['Threshold'], df[col], label=col)
    plt.xlabel('Threshold')
    plt.ylabel('Score')
    plt.title('M√©tricas vs Threshold (CV en TRAIN)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("grafico_thresholds.png")
    plt.close()
    return df
```

```python
def evaluar_en_test(modelo, X_test, y_test, label_enc, threshold=0.4, beta=2.0,
                    modelo_nombre="modelo", carpeta="resultados"):
    """Eval√∫a el modelo en TEST y guarda los resultados con formato profesional."""

    # === Crear carpeta si no existe ===
    os.makedirs(carpeta, exist_ok=True)

    logger.info("\nüìà Evaluando modelo en conjunto de TEST...")
    y_probs = modelo.predict_proba(X_test)[:, 1]
    y_pred = (y_probs >= threshold).astype(int)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    f2 = fbeta_score(y_test, y_pred, beta=beta)
    auc = roc_auc_score(y_test, y_probs)

    logger.info("ROC-AUC: %.3f", auc)
    logger.info("Threshold: %.2f", threshold)
    logger.info("Precision: %.3f", precision)
    logger.info("Recall: %.3f", recall)
    logger.info("F1-score: %.3f", f1)
    logger.info("F%s-score: %.3f", int(beta), f2)

    # === Crear formato profesional ===
    fecha_actual = datetime.now().strftime("%Y-%m-%d")

    contenido = (
        "üìå Resultados finales del modelo\n\n"
        f"ROC-AUC: {auc:.3f}\n"
        f"Threshold √≥ptimo: {threshold:.2f}\n\n"
        "üìä M√©tricas en TEST\n"
        f"Precision: {precision:.3f}\n"
        f"Recall: {recall:.3f}\n"
        f"F1-score: {f1:.3f}\n"
        f"F2-score: {f2:.3f}\n\n"
        f"Fecha de ejecuci√≥n: {fecha_actual}\n"
    )

    # === Guardar archivo ===
    txt_path = f"{carpeta}/threshold_optimo_y_metricas.txt"
    with open(txt_path, "w", encoding="utf-8") as f:
        f.write(contenido)

    logger.info("üìù Archivo guardado en: %s", txt_path)

    # === Matriz de confusi√≥n ===
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_enc.classes_)
    disp.plot(cmap="Blues")
    plt.title("Matriz de confusi√≥n (Test)")
    plt.tight_layout()
    plt.savefig(f"{carpeta}/matriz_confusion_{modelo_nombre}.png")
    plt.close()


def analizar_probabilidades(modelo, X_test, y_test_enc, threshold=0.4, modelo_nombre="modelo", label_enc=None):
    """Analiza la distribuci√≥n de probabilidades en test."""
    y_probs = modelo.predict_proba(X_test)[:, 1]
    y_preds = (y_probs >= threshold).astype(int)

    df_probs = pd.DataFrame({
        'Probabilidad_Churn': y_probs,
        'Predicci√≥n': y_preds,
        'Real': y_test_enc
    })

    if label_enc:
        df_probs['Real_label'] = label_enc.inverse_transform(df_probs['Real'])
        df_probs['Predicci√≥n_label'] = label_enc.inverse_transform(df_probs['Predicci√≥n'])
    else:
        df_probs['Real_label'] = df_probs['Real'].map({0: 'Clase 0', 1: 'Clase 1'})
        df_probs['Predicci√≥n_label'] = df_probs['Predicci√≥n'].map({0: 'Clase 0', 1: 'Clase 1'})

    plt.figure(figsize=(10, 6))
    plt.hist(df_probs[df_probs['Real'] == 1]['Probabilidad_Churn'], bins=25, alpha=0.6, label='Churn = Yes')
    plt.hist(df_probs[df_probs['Real'] == 0]['Probabilidad_Churn'], bins=25, alpha=0.6, label='Churn = No')
    plt.axvline(x=threshold, color='red', linestyle='--', label=f'Umbral {threshold}')
    plt.xlabel('Probabilidad de Churn')
    plt.ylabel('N√∫mero de clientes')
    plt.title('Distribuci√≥n de Probabilidades por Clase Real')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"distribucion_probabilidades_{modelo_nombre}.png")
    plt.close()
```

```python
# ================== MAIN ==================
def main(dataset_path, fs_threshold='median', use_fs=True, umbral_manual=None):
    """Ejecuta el flujo completo de entrenamiento, evaluaci√≥n y guardado del modelo."""

    # üìÅ Crear carpeta de resultados para este experimento
    carpeta = f"resultados_{fs_threshold}"
    os.makedirs(carpeta, exist_ok=True)

    # 1. Cargar y preprocesar datos
    X, y = cargar_y_preprocesar_datos(dataset_path)

    # 2. Dividir datos
    X_train, X_test, y_train_enc, y_test_enc, label_enc = dividir_datos(X, y)

    # 3. Crear preprocesador
    preprocessor = construir_preprocesador(X)

    # 4. Configurar validaci√≥n cruzada
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # 5. Entrenar y ajustar modelos
    resultados = entrenar_modelos(X_train, y_train_enc, preprocessor, cv, fs_threshold, force_fs=use_fs)

    # 6. Seleccionar mejor modelo
    mejor_tag = max(resultados, key=lambda t: resultados[t]['score'])
    best_model = resultados[mejor_tag]['model']
    logger.info("\nüèÜ Mejor pipeline: %s con recall CV = %.4f", mejor_tag, resultados[mejor_tag]['score'])

    # 6.1 Interpretabilidad con SHAP
    plt.figure()
    explicar_modelo_final_shap(best_model, X_test, modelo_nombre=f"{mejor_tag}_{fs_threshold}")
    try:
        os.replace(
            f"shap_summary_{mejor_tag}_{fs_threshold}_final.png",
            f"{carpeta}/shap_summary_{fs_threshold}.png"
        )
    except Exception as e:
        logger.warning("No se pudo mover el archivo SHAP (puede que no exista): %s", e)

    # 7. Mostrar y guardar features seleccionadas
    try:
        mostrar_features_seleccionados(best_model, X_train, f"{mejor_tag}_{fs_threshold}")
        os.replace(
            f"features_seleccionadas_{mejor_tag}_{fs_threshold}.csv",
            f"{carpeta}/features_seleccionadas_{fs_threshold}.csv"
        )
    except Exception as e:
        logger.warning("Error al guardar/mostrar features seleccionadas: %s", e)

    # 8. Ajustar threshold
    tabla_metricas = ajustar_threshold(best_model, X_train, y_train_enc)
    tabla_path = f"{carpeta}/tabla_thresholds_{fs_threshold}.csv"

    try:
        os.replace("tabla_thresholds.csv", tabla_path)
        os.replace("grafico_thresholds.png", f"{carpeta}/grafico_thresholds_{fs_threshold}.png")
    except Exception as e:
        logger.warning("No se pudieron mover archivos de thresholds/grafico: %s", e)

    # 9. Selecci√≥n del threshold final
    threshold_optimo = tabla_metricas.loc[tabla_metricas['F2-score'].idxmax(), 'Threshold']

    if umbral_manual is not None and 0 <= umbral_manual <= 1:
        threshold_final = umbral_manual
        logger.info("‚öôÔ∏è Usando umbral personalizado definido por el usuario: %s", threshold_final)
    else:
        threshold_final = threshold_optimo
        logger.info("‚úÖ Usando umbral √≥ptimo por F2-score: %.2f", threshold_final)

    logger.info("‚û°Ô∏è Threshold final para predicci√≥n: %.2f", threshold_final)

    # 10. Evaluar en test
    evaluar_en_test(
        best_model, X_test, y_test_enc, label_enc,
        threshold=threshold_final,
        modelo_nombre=f"{mejor_tag}_{fs_threshold}",
        carpeta=carpeta
    )

    # 11. Analizar probabilidades
    analizar_probabilidades(
        best_model, X_test, y_test_enc,
        threshold=threshold_final,
        modelo_nombre=f"{mejor_tag}_{fs_threshold}",
        label_enc=label_enc
    )
    try:
        os.replace(
            f"distribucion_probabilidades_{mejor_tag}_{fs_threshold}.png",
            f"{carpeta}/distribucion_probabilidades_{fs_threshold}.png"
        )
    except Exception as e:
        logger.warning("No se pudo mover la distribucion de probabilidades: %s", e)

    # 12. Guardar modelo y threshold
    model_path = f"{carpeta}/mejor_modelo_{fs_threshold}.pkl"
    joblib.dump(best_model, model_path)

    logger.info("‚úÖ Modelo y threshold guardados en %s", carpeta)
    logger.info("\n‚öôÔ∏è Par√°metros activos ‚Üí fs_threshold='%s', umbral_manual=%s\n", fs_threshold, umbral_manual)


# ================== EJECUCI√ìN ==================
if __name__ == '__main__':
    DATASET_PATH = 'WA_Fn-UseC_-Telco-Customer-Churn.csv'

    # Ejecutar para ambas configuraciones
    FS_OPTIONS = ['mean', 'median']
    UMBRAL_MANUAL = 0.4  # o None para usar el √≥ptimo

    for fs in FS_OPTIONS:
        logger.info("\n===================== EJECUTANDO fs_threshold = %s =====================", fs)
        main(DATASET_PATH, fs_threshold=fs, use_fs=True, umbral_manual=UMBRAL_MANUAL)
```
