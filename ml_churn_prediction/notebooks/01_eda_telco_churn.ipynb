# Exploratory Data Analysis (EDA)

Este notebook explora el dataset de churn para comprender la estructura de los datos,
identificar valores faltantes y observar distribuciones relevantes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Cargar dataset

```python
df = pd.read_csv("../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

## Información general del dataset

```python
df.info()
```

## Tamaño del dataset y columnas

```python
print("Shape del dataset:", df.shape)
print("\nColumnas:")
print(df.columns.tolist())
```

## Valores nulos

```python
df.isnull().sum()
```

## Distribución del target (Churn)

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Churn", data=df)

plt.title("Distribución de Churn")
plt.show()
```

## Distribución de variables numéricas

```python
num_cols = df.select_dtypes(include=np.number).columns

df[num_cols].hist(figsize=(12,8), bins=30)
plt.tight_layout()
plt.show()
```

## Variables categóricas

```python
cat_cols = df.select_dtypes(include="object").columns

cat_cols
```

## Ejemplo de análisis categórico

```python
plt.figure(figsize=(8,4))
sns.countplot(x="InternetService", hue="Churn", data=df)

plt.title("InternetService vs Churn")
plt.xticks(rotation=30)
plt.show()
```
